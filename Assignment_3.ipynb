```python
# Method to find beta values
def CalcBeta(X,Y):
    A = X.T.dot(X)
    B = np.linalg.inv(A)
    C = B.dot(X.T)
    beta = C.dot(Y)
    return beta

# Method to get predicted Y values
def GetPred(X,B):
    Y = X.dot(B)
    return Y

# Method to calculate R2 score
def R2(Y,Y2):
    error = Y2 - Y
    error_square = np.power(error,2)
    sum_square_error = np.sum(error_square)
    mse = sum_square_error / len(Y2)
    rmse = np.sqrt(mse)
    y_mean = np.mean(Y)
    var = np.sum((Y-y_mean)**2)
    r2 = 1-(sum_square_error/var)
    return r2
```

```python
# Converting values into array format
x_train = np.array(x_train)
y_train = np.array(y_train)
x_valid = np.array(x_valid)
y_valid = np.array(y_valid)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
# Function to perform MLR
def train_model(rate,iter,x_train,y_train):
    beta=np.zeros(6)
    number_of_iterations=iter
    learning_rate=rate
    n = x_train.shape[0]
    for i in range(number_of_iterations):
        x0_gradient=0
        x1_gradient=0
        x2_gradient=0
        x3_gradient=0
        x4_gradient=0
        x5_gradient=0    
        for j in range(len(x_train)):
            a=x_train[j,0]
            b=x_train[j,1]
            c=x_train[j,2]
            d=x_train[j,3]
            e=x_train[j,4]
            f=y_train[j]
            x0_gradient+=(beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)
            x1_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*a)
            x2_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*b)
            x3_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*c)
            x4_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*d)
            x5_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*e)
        beta[0]=beta[0]-learning_rate/n*x0_gradient
        beta[1]=beta[1]-learning_rate/n*x1_gradient
        beta[2]=beta[2]-learning_rate/n*x2_gradient
        beta[3]=beta[3]-learning_rate/n*x3_gradient
        beta[4]=beta[4]-learning_rate/n*x4_gradient
        beta[5]=beta[5]-learning_rate/n*x5_gradient
        return beta
```

```python
# Function to get predicted values
def pred(x_test,beta):
    #x_test=np.insert(x_test,0,values=1,axis=1)
    y_predict=x_test.dot(beta)
    return y_predict
```

```python
# Function to calculate R2 Score
def eval(y_test,Y_predict):
    error=y_test-y_predict
    square_error=np.power(error,2)
    sum_square_error=np.sum(square_error)
    mean_square_error=sum_square_error/len(y_predict)
    #print(mean_square_error)
    rms_error=np.sqrt(mean_square_error)
    #print(rms_error)
    y_mean=np.mean(y_test)
    total_variance=np.sum((y_test-y_mean)**2)
    #print('R2 Score:',1-sum_square_error/total_variance)
    return (1-sum_square_error/total_variance)
```

```python
# Driver Code
beta = []
r2 = []
beat_temp = train_model(0.001,1000,x_train,y_train)
pred_temp = pred(x_test,beta_temp)
r2_temp = eval(y_test,pred_temp)
print('LR = 0.001     |  R2 =',r2_temp)
beta.append(beta_temp)
r2.append(r2_temp)
beat_temp = train_model(0.01,1000,x_train,y_train)
pred_temp = pred(x_test,beta_temp)
r2_temp = eval(y_test,pred_temp)
print('LR = 0.01      |  R2 =',r2_temp)
beta.append(beta_temp)
r2.append(r2_temp)
beat_temp = train_model(0.1,1000,x_train,y_train)
pred_temp = pred(x_test,beta_temp)
r2_temp = eval(y_test,pred_temp)
print('LR = 0.1       |  R2 =',r2_temp)
beta.append(beta_temp)
r2.append(r2_temp)
beat_temp = train_model(0.001,1000,x_train,y_train)
pred_temp = pred(x_test,beta_temp)
r2_temp = eval(y_test,pred_temp)
print('LR = 1         |  R2 =',r2_temp)
beta.append(beta_temp)
r2.append(r2_temp)
```
